```python
import regex as re
import pandas as pd
```

```python
def clean_text(text):
    # Menghapus tanda baca dan karakter non-alfanumerik
    cleaned_text = re.sub(r'[^a-zA-Z\s]', '', text)
    return cleaned_text
```

```python
input_file = "result.csv"
output_file = "cleaned_result.csv"
```

# Soal 3d

d.Lakukan stopword removing  berdasarkan hasil 3c, simpan hasilnya dalam bentuk csv/excel. (bobot:10)!

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download dataset stopwords untuk bahasa Indonesia
nltk.download('stopwords')
nltk.download('punkt')
```

```python
# Baca file CSV dengan dataset Twitter (tokenized_twitter_data.csv)
twitter_data = pd.read_csv('tokenized_twitter_data.csv', encoding='utf-8')

# Fungsi untuk menghapus stopwords
def remove_stopwords(text):
    stop_words = set(stopwords.words('indonesian'))
    words = word_tokenize(text)
    return [word for word in words if word.lower() not in stop_words]

# Melakukan penghapusan stopwords pada kolom "Tokens"
twitter_data['Tokens'] = twitter_data['Tokens'].apply(remove_stopwords)

# Simpan hasilnya dalam file CSV baru
twitter_data.to_csv('stopword_twitter_data.csv', index=False, encoding='utf-8')

print('Stopwords telah dihapus dan hasilnya disimpan dalam "stopword_twitter_data.csv".')
```
