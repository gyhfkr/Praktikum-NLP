```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# 1. Load dataset "data.csv" menjadi dataframe
data = pd.read_csv("data.csv")
```

```python
data.head()
```

```python
# 2. Ambil 5 data dan ubah kolom "content" menjadi vektor TF-IDF
sample_data = data.head(5)  # Ambil 5 data pertama
content = sample_data["content"]
```

```python
# Inisialisasi TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Transformasi teks menjadi vektor TF-IDF
tfidf_matrix = tfidf_vectorizer.fit_transform(content)
```

```python
# Tampilkan hasilnya
print("Hasil TF-IDF Vector untuk 5 data pertama:")
print(tfidf_matrix)
```

```python
# 3. Tampilkan nama fiturnya dengan get_feature_names_out dan Compute TF-IDF
feature_names = tfidf_vectorizer.get_feature_names_out()
print("\nNama fitur:")
print(feature_names)
```

```python
# Compute Term Frequency (TF) dan Inverted Document Frequency (IDF)
for i, data_idx in enumerate(sample_data.index):
    print(f"\nData ke-{i + 1}:")
    data_tf = tfidf_matrix[i].toarray()  # Mendapatkan vektor TF-IDF untuk data tertentu
    for j, feature_name in enumerate(feature_names):
        print(f"{feature_name}: {data_tf[0][j]:.4f}")
```

```python
# Buat dataframe dari matriks TF-IDF dengan nama kolom sesuai dengan fitur kata
tfidf_df = pd.DataFrame(data=tfidf_matrix.toarray(), columns=feature_names)

# Tampilkan matriks kata
print(tfidf_df)
```
